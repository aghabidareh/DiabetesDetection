```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.impute import KNNImputer
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

from sklearn.metrics import classification_report
from catboost import CatBoostClassifier

from imblearn.over_sampling import RandomOverSampler
```

```python
data = pd.read_csv('diabetes_prediction_dataset.csv')
```

```python
x = data.iloc[: , :-1]
y = data.iloc[: , -1]
```

```python
le = LabelEncoder()
x.iloc[: , 0] = le.fit_transform(x.iloc[: , 0])
x.iloc[: , 4] = le.fit_transform(x.iloc[: , 4])
```

```python
x.replace({'smoking_history':0} , np.nan , inplace=True)
```

```python
imputer = KNNImputer(n_neighbors=2 , weights='distance')
scaler = StandardScaler()
minmax = MinMaxScaler(clip=True)
x = imputer.fit_transform(x)
x = scaler.fit_transform(x)
x = minmax.fit_transform(x)
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , random_state=1441 , test_size=0.2 , train_size=0.8)
```

```python
rus = RandomOverSampler()
x_train , y_train = rus.fit_resample(x_train , y_train)
```

```python
model = CatBoostClassifier(random_state=1441 , iterations=10000 , learning_rate=0.001, verbose=2000 , depth=8 , eval_metric='Recall' , task_type='GPU')
model.fit(x_train , y_train, eval_set=(x_test , y_test))
preds = model.predict(x_test)
print(classification_report(y_test , preds))
```
